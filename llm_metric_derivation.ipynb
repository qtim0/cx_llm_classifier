```python
import pandas as pd
import numpy as np
import sklearn
import cfunc
```

```python
raw = pd.read_csv('gpt_classified.csv')
raw
```

```python
# Sample of doing it step by step, the cfunc.py library has a function that does this now.

cat = 'DELIVERY'

tdf = raw
tdf = tdf[(tdf['chatgpt_class'] == cat) | (tdf['category'] == cat)]
print('n: ' + str(len(tdf)))

accuracy = round(sum(tdf['correct']) / len(tdf['correct']) * 100, 2)
precision = round(sum(tdf['correct']) / len(tdf['chatgpt_class']) * 100, 2)
recall = round(sum(tdf['correct']) / len(tdf[tdf['category'] == cat]) * 100, 2)
f1 = round((2 * precision * recall) / (precision + recall), 2)
print('accuracy:' + str(accuracy) + '\n' +
        'precision: ' + str(precision) + '\n' +
        'recall: ' + str(recall) + '\n' +
        'f1: ' + str(f1))

tdf
```

```python
wdf = raw

payment_df = cfunc.per_category_metric(data = wdf, category = 'PAYMENT')
delivery_df = cfunc.per_category_metric(data = wdf, category = 'DELIVERY')
invoice_df = cfunc.per_category_metric(data = wdf, category = 'INVOICE')
account_df = cfunc.per_category_metric(data = wdf, category = 'ACCOUNT')
order_df = cfunc.per_category_metric(data = wdf, category = 'ORDER')
cancel_df = cfunc.per_category_metric(data = wdf, category = 'CANCEL')
contact_df = cfunc.per_category_metric(data = wdf, category = 'CONTACT')
refund_df = cfunc.per_category_metric(data = wdf, category = 'REFUND')
#subscription_df = cfunc.per_category_metric(data = wdf, category = 'SUBSCRIPTION')
#update_df = cfunc.per_category_metric(data = wdf, category = 'UPDATE')

mdf = pd.concat([payment_df, delivery_df, invoice_df, account_df, order_df, cancel_df, contact_df, refund_df])
mdf
```

```python
macros = mdf.groupby('metric')['value'].mean().reset_index()
macros
```
